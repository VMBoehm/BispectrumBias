```python
%matplotlib inline
import matplotlib.pyplot as plt
import pickle
import numpy as np
plt.style.use(['seaborn-colorblind','paper'])
plt.style.use('classic')
```

```python
ellpath="/home/nessa/Documents/Projects/LensingBispectrum/CMB-nonlinear/outputs/ells/"
specpath='/home/nessa/Documents/Projects/LensingBispectrum/CMB-nonlinear/outputs/power_spectra/'
bispecpath='/home/nessa/Documents/Projects/LensingBispectrum/CMB-nonlinear/outputs/bispectra/'
```

```python
#class Pk with and without massive neutrinos

kk, pk1 = pickle.load(open(specpath+'Pk_test1_massive_nus_equilat_JiaGalaxyLens_halofit.pkl','r'))
kk, pk2 = pickle.load(open(specpath+'Pk_test1_equilat_JiaGalaxyLens_halofit.pkl','r'))


f_nu=0.1/93.14/0.14 

plt.semilogx(kk,pk1/pk2)
plt.axhline((1.-8.*f_nu),ls='--') #should be approx. suppression
```

```python
#compare class cls and 

L,l,theta = pickle.load(open(ellpath+'ell_ang_equilat_Lmin1_Lmax3000_lmin1_lmax8000_lenL160_lenl180_lenang160_1e-04.pkl','r'))
ll, cls1t, cls2t = np.load(specpath+'CL_test1_equilat_JiaGalaxyLens_Lmin1-Lmax2999_halofit_CLASS.npy')
ll, cls1_, cls2_ = np.load(specpath+'CL_test1_equilat_JiaGalaxyLens_Lmin1-Lmax2999__CLASS.npy')
cls1 = np.load(specpath+'CL_test1_equilat_JiaGalaxyLens_Lmin1-Lmax2999_halofit.npy')

plt.figure()
plt.loglog(ll,cls1t) #full sky
plt.loglog(ll,cls2t) #flat sky
plt.loglog(ll,cls1_,':') #linear
plt.loglog(L,cls1,ls='--')
plt.xlim(10,5000)
plt.show()

plt.figure()
plt.semilogx(L,np.interp(L,ll,cls2t)/cls1-1)
plt.ylim(-0.1,0.1)
plt.xlim(100,3000)
```

```python
#same thing including neutrinos

L,l,theta = pickle.load(open(ellpath+'ell_ang_equilat_Lmin1_Lmax3000_lmin1_lmax8000_lenL160_lenl180_lenang160_1e-04.pkl','r'))
ll, cls1t, cls2t = np.load(specpath+'CL_test1_massive_nus_equilat_JiaGalaxyLens_Lmin1-Lmax2999_halofit_CLASS.npy')
cls1 = np.load(specpath+'CL_test1_massive_nus_equilat_JiaGalaxyLens_Lmin1-Lmax2999_halofit.npy')

plt.figure()
plt.loglog(ll,cls1t) #full sky
plt.loglog(ll,cls2t) #flat sky
plt.loglog(L,cls1,ls='--')
plt.xlim(10,5000)
plt.show()

plt.figure()
plt.semilogx(L,np.interp(L,ll,cls2t)/cls1-1)
plt.ylim(-0.1,0.1)
plt.xlim(100,3000)
```

```python
bi2 = np.load(bispecpath+'bispec_test1_massive_nus_equilat_JiaGalaxyLens_Lmin1-Lmax2999-lmax2999_halofit_.npy')
bi1 = np.load(bispecpath+'bispec_test1_equilat_JiaGalaxyLens_Lmin1-Lmax2999-lmax2999_halofit_.npy')
plt.figure()
plt.semilogx(L,bi2/bi1)
plt.axhline((1.-0.065)**2,ls='--') #expectation from 1-f_nu suppression
plt.axhline((1.-0.085)**(3./2.),ls=':') #expectation of supression of Cls if I assume B~Cl^(3/2)
plt.axhline((1.-0.085)**2,ls='-.') #expectation of supression of Cls if I assume B~Cl^2
plt.xlim(100,3000)
```
