```python
name='CL_Cltest17'
L,l,theta = pickle.load(open(ellpath+'ell_ang_equilat_Lmin1_Lmax3000_lmin1_lmax8000_lenL200_lenl220_lenang200_1e-04.pkl','r'))

CLme = np.load(specpath+'%s_equilat_JiaGalaxyLens_Lmin1-Lmax2999_.npy'%name)
ll, CLCLASS1, CLCLASS2 = np.load(specpath+'%s_equilat_JiaGalaxyLens_Lmin1-Lmax2999__CLASS.npy'%name)

plt.figure()
plt.semilogx(L,(CLme/np.interp(L,ll,CLCLASS1))-1, label='1')
plt.semilogx(L,(CLme/np.interp(L,ll,CLCLASS2))-1, label='2')
plt.semilogx(L,(CLme*L**4/(L+0.5)**4/np.interp(L,ll,CLCLASS2))-1, label='3', ls='--')
plt.axhline(y=0.001,ls='--',color='black',label='0.1%')
plt.legend()
plt.xlabel('L')
plt.ylabel('$C_{L_{Vanessa}}/C_{L_{CLASS}}-1$',fontsize=20)
plt.xlim(100,3000)
plt.ylim(0.00,0.02)

plt.grid()
plotpath='/home/nessa/Documents/Projects/LensingBispectrum/CMB-nonlinear/Tests/GalaxyLensing/Cls/'
plt.savefig(plotpath+'%s.png'%name,bbox_inches='tight')
```
